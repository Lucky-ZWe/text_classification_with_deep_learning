```python
"""计数统计"""
import pandas as pd
from pandas import Series

train_filename = "../../data/sogou_resource/dataset/extract_train.csv.zip"
test_filename = "../../data/sogou_resource/dataset/extract_test.csv.zip"

df_train = pd.read_csv(train_filename,compression='zip')
# counts = df_train['type'].value_counts()
# print counts

df_test = pd.read_csv(test_filename,compression='zip')
# test_counts = df_test['type'].value_counts()
# print test_counts
content_df = df_train.append(df_test,ignore_index=True)
counts = content_df['type'].value_counts()

# 绘图
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
counts[:].plot('bar')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np
# y_true = ['cat', 'ant', 'cat', 'cat', 'ant', 'bird']
# y_pred = ['ant', 'ant', 'cat', 'cat', 'ant', 'cat']
y_true = [1,2,3,2,1,3,2]
y_pred = [1,2,2,2,1,3,2]
labels = {1:'ant', 2:'bird',3:'cat'}
y_true = [labels[x] for x in y_true]
y_pred = [labels[x] for x in y_pred]
maxtrix = confusion_matrix(y_true, y_pred,labels=labels.values())
plt.matshow(maxtrix)
plt.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
plt.colorbar()
plt.xlabel('Prediction')
plt.ylabel('True')
plt.xticks(np.arange(maxtrix.shape[1]),labels.values())
plt.yticks(np.arange(maxtrix.shape[1]),labels.values())
plt.show()
```
